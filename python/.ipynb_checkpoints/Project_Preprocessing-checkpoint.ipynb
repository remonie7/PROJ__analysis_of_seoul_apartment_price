```python
a=pd.read_csv("real.csv")
a
```

```python
floor = []
totalfloor = []

for i in range(0,len(a)):
    floor.append(a["층"][i].split("/")[0])
    totalfloor.append(float(a["층"][i].split("/")[1]))
```

```python
a["현재층"]=floor
a["전체층"]=totalfloor
a
```

```python
floorval=[]
```

```python
#a["현재층"].value_counts()

for i in range(0,len(a)):
    if a["현재층"][i]=="고": floorval.append(4)
    elif a["현재층"][i]=="중": floorval.append(3)
    elif a["현재층"][i]=="저": floorval.append(2)
    elif a["현재층"][i]=="1" : floorval.append(1)
    else : 
        fl = float(a["현재층"][i])/float(a["전체층"][i])
        if fl==1: floorval.append(5)
        elif fl>=0.7 : floorval.append(4)
        elif fl>=0.4 : floorval.append(3)
        else : floorval.append(2)
```

```python
a["현재층범주"]=floorval
```

```python
a["현재층범주"].value_counts()
```

```python
#a.drop(["Unnamed: 0"], axis=1, inplace=True)
a.drop(["층"], axis=1, inplace=True)
a.drop(["현재층"], axis=1, inplace=True)
a.drop(["전체층"], axis=1, inplace=True)
```

```python
a.to_csv("finaldata.csv")
```
