```python
total = infomation['total']
total_page=0;
if total%20 ==0:
    total_page=total//50
else:
    total_page=total//50 + 1
```

```python
def get_api_result(keyword, display, start):
    url = "https://openapi.naver.com/v1/search/blog?query=" + keyword + "&display=" \
        + str(display) + "&start=" + str(start) 
    result=requests.get(url, headers=headers)
    return result.json()
```

```python
def call_and_print(keyword, page):
    json_object = get_api_result(keyword, 50, (page-1)*50+1)
    for item in json_object['items']:
        title = item['title'].replace("<b>","").replace("</b>","").replace("'","&quot;")
        description = item['description'].replace("<b>","").replace("</b>","").replace("'","&quot;")
        bloggername = item['bloggername'].replace("'","&quot;")
        link = item['link']
        print(title, description, bloggername, link)
        
        insert_sql = "insert into blog_info values(blog_seq.nextval, '{}', '{}', '{}', '{}')".format(title, description, bloggername, link)
        cursor.execute(insert_sql)
        connection.commit()
```

```python
keyword="역삼역"

for page in range(1, total_page):
    print(page, " page")
    call_and_print(keyword, page)
    time.sleep(0.5)
```

```python
blog_df=pd.read_sql("select * from blog_info", connection)
blog_df.head()
```

```python
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgunbd.ttf").get_name()
rc('font', family=font_name)
```

```python
pattern="[^\w\s]" # [^패턴] : 패턴에 맞는 문자만 리턴, \w : 숫자 영어 한글, \s:공백
#re.sub(pattern, 패턴에 맞지않는 문자를 바꿀문자, 원분에서 패턴에 맞지않는 문자는 다른 문자로 변환. 원본에서 패턴에 맞는건 그대로 리턴)
blog_text = [re.sub(pattern=pattern, repl='', string=text).replace("강남","").replace("역삼","").replace("역","").replace("quot","").replace("ㅡ", "") for text in blog_df["DESCRIPTION"]] 
#repl 속성은 숫자영어한글 아닌거 변환리턴값
blog_text
```

```python

tag_list = []
for text in blog_text:
    tagged = twitter.pos(text, stem=True)
    #print(tagged)
    for i in range(0, len(tagged)):
        if(tagged[i][1]=="Noun") | (tagged[i][1]=="Adjective"):
            tag_list.append(tagged[i][0])
        
tag_list
```

```python
tag_list = [tag for tag in tag_list if len(tag)>=2]
tag_list
```

```python
ko = nltk.Text(tag_list)
ko.vocab().most_common(100)
```

```python
plt.figure(figsize=(15,6))
ko.plot(50)
plt.show()
```

```python
data = ko.vocab().most_common(300)

wordcloud = WordCloud(font_path='c:/Windows/Fonts/malgunbd.ttf', relative_scaling=0.2, stopwords=STOPWORDS, background_color='white',).generate_from_frequencies(dict(data))
plt.figure(figsize=(16,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
